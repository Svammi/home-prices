Рассмотрим распределение цен по каждому классу зданий и найдем среднюю стоимость здания определенного класса.

```python
# получим новую таблицу, где сопоставим цены классу здания
price = pd.read_csv("house-prices-advanced-regression-techniques/sample_submission.csv")
```

```python
price_and_info_about_home = pd.merge(test_data, price, on='Id')
price_and_sub_class = pd.merge(test_data, price, on='Id').loc[:, ['Id', 'MSSubClass', 'SalePrice']]
price_and_sub_class_sort_by_class = {class_: price_and_sub_class[price_and_sub_class.MSSubClass == class_] for class_ in ms_sub_class}
print(price_and_sub_class)
```

```python
# найдем модальную цену для непрерывной случайной величины определенного класса
def modal_value_for_interval_variation_series(num_class, mas):
    x = len(mas[num_class]['SalePrice']) 
    # число частичных интервалов
    count_interval = int(1 + 3.322*math.log(x, 10)) 
    h =( max(mas[num_class]['SalePrice']) - min(mas[num_class]['SalePrice']) )/ (count_interval+1)
    
    hist, bins = np.histogram(mas[num_class]['SalePrice'], bins = count_interval+1)
        
    # получим средние значения каждого из интервалов
    average_interval_values = [bins[i]+ (bins[i+1] - bins[i])/2 for i in range(len(bins)-1)] 
    
    interval_variation_series = pd.DataFrame({'IVS_0':bins[:-1], 'IVS_1':bins[1:], 'Average':average_interval_values, 'Count':hist})
        
    # найдем модальный интервал
    modal_interval = interval_variation_series[interval_variation_series.Count == max(interval_variation_series.Count)]
    moda = []
    # найдем модальное значение для интервального вариационного ряда
    for position in modal_interval.index.values:
        #position = modal_interval.index[0] 

        if position != 0 and position != interval_variation_series.index[-1]:
            n_1 = modal_interval['Count'][position] - interval_variation_series['Count'][position-1]
            n_2 = modal_interval['Count'][position] - interval_variation_series['Count'][position+1]
        elif position == 0:
            n_1 = 0
            n_2 = modal_interval['Count'][position] - interval_variation_series['Count'][position+1]
        else:
            n_1 = modal_interval['Count'][position] - interval_variation_series['Count'][position-1]
            n_2 = 0

        moda.append(modal_interval['IVS_0'] + (n_1)/(n_1 + n_2) * h)
    
    plt.figure(figsize=(10,5))
    plt.title("Frequrency prices for klass" + str(num_class))
    plt.xlabel("Price")
    plt.hist(mas[num_class]['SalePrice'], bins=bins)

    
    
    return moda
```

```python

for i in price_and_sub_class_sort_by_class.keys():
    print(i)
    moda = modal_value_for_interval_variation_series(i,price_and_sub_class_sort_by_class)
    print(moda)

```

Попарная корреляция для всех параметров таблицы

```python
colormap = plt.cm.viridis
plt.figure(figsize=(80,80))
plt.titile = ("Попарная корреляция")
sbn.heatmap(price_and_info_about_home.corr(), square=True, annot=True, cmap=colormap)
plt.show()
```

```python
# получим паарметры с наибольшим значениям коэффициента корреляции
df_corr = price_and_info_about_home.corr().unstack().reset_index() # строит olap куб 
hight_corr_features = df_corr[(np.abs(df_corr[0])>0.75) & (df_corr['level_0'] != df_corr['level_1'])]
print(hight_corr_features)
```

```python
# построим диаграммы рассеяния для каждой коррелирующей пары
for feature_pair in hight_corr_features.index:
    plt.figure(figsize=(16,3))
    plt.plot(price_and_info_about_home[hight_corr_features.loc[feature_pair, 'level_0']], price_and_info_about_home[hight_corr_features.loc[feature_pair, 'level_1']], 'o', alpha=0.7)
    plt.title('Correlation %f' % round(hight_corr_features.loc[feature_pair, 0],4))
    plt.xlabel(hight_corr_features.loc[feature_pair, 'level_0'])
    plt.ylabel(hight_corr_features.loc[feature_pair, 'level_1'])
    plt.show()
```
